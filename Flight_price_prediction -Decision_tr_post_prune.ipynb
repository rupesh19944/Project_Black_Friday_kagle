```python
df_test = df[df['Price'].isnull()]
df_train = df[~df['Price'].isnull()]
```

```python
X = df_train.loc[:,~df_train.columns.isin(['Price'])]
y= df_train['Price']
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.describe()
```

##### Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor(random_state=0)
dt.fit(X_train,y_train)
print(dt.score(X_train,y_train))
dt.score(X_test,y_test)
```

```python
dt.feature_importances_
```

```python
from sklearn.metrics import mean_squared_error
```

```python
dt.tree_.max_depth
```

```python
# running up to 16 depth 

mean_sq_error = []
for depth in range(1, (dt.tree_.max_depth +1)):
    d_tree_reg = DecisionTreeRegressor(max_depth=depth)
    d_tree_reg.fit(X_train, y_train)
    tree_predictions = d_tree_reg.predict(X_test)
    mean_sq_error.append(mean_squared_error(y_test, tree_predictions))

tree_depths = [depth for depth in range(1, (dt.tree_.max_depth +1))]
plt.figure(figsize=(10,  6))
plt.grid()
plt.plot(tree_depths, mean_sq_error)
plt.xlabel("Tree Depth")
plt.ylabel("Mean Square Error")
```

###### Observation: best depth for the tree can bw 12 to 19

```python
best_param_model = DecisionTreeRegressor(max_depth=19, random_state=42, min_samples_leaf=6,)
```
